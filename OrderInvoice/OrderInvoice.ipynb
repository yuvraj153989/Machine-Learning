**APPROACH**

we will represent all value counts within a specific ranges, and store them into temporary column df["UPDATED_DISTRIBUTION_CHANNEL"](which we will delete later)

```python
value_counts_dict = {}
for value in df["DISTRIBUTION_CHANNEL"].unique():
    value_counts = df["DISTRIBUTION_CHANNEL"].value_counts()[value]
    value_counts_dict[value] = value_counts
print(value_counts_dict)

#storing every unique values in dictionary with key and values,
#where key represents Country/City/State name, and values represnt their respective distribution
```

```python
# occurrence ranges and corresponding labels, using tuples
ranges = [(0, 4300, "Below 4300"),(4300, 4350, "4300-4350"),(4350, 4400, "4350-4400"),(4400, 4450, "4400-4450"),
          (4450, 4500, "4450-4500"),(4500, 4550, "4500-4550"),(4550, 4600, "4550-4600"),(4600, 4650, "4600-4650")
         ,(4650, 4700, "4650-4700"),(4700, 4750, "4700-4750"),(4750, 4800, "4750-4800"),(4800, 4850, "4800-4850")
         ,(4850, 4900, "4850-4900"),(4900, 4950, "4900-4950"),(4950, 5000, "4950-5000")]
# Create the UPDATED_DISTRIBUTION_CHANNEL column using the value_counts_dict (above created dictionary)
df["UPDATED_DISTRIBUTION_CHANNEL"] = df["DISTRIBUTION_CHANNEL"].map(value_counts_dict).apply(
    lambda x: next((label for start, end, label in ranges if start <= x <= end), "Above 5000"))
# Verifying the changes
print(df["UPDATED_DISTRIBUTION_CHANNEL"].value_counts())
```

```python
import datetime as dt
def calculate_time_features(df, start, end, date_col, customer_col, sales_col):
    # Convert the date column to datetime format
    df[date_col] = pd.to_datetime(df[date_col])
    
    # Sort the DataFrame by the date column in ascending order
    df = df.sort_values(date_col)
    
    # Create empty columns for the time features
    df['Last_day_sales'] = 0
    df['Last_day_diff'] = 0
    df['Year'] = 0
    df['Month'] = 0
    df['Week'] = 0
    df['Day'] = 0
    df['Day_of_week'] = 0
    df['Day_of_year'] = 0
    
    for i in range(start, end+1):
        # Calculate last day sales
        df['Last_day_sales'] = df.groupby(customer_col)[sales_col].shift(i).fillna(0)
        
        # Calculate last day difference
        df['Last_day_diff'] = df.groupby(customer_col)['Last_day_sales'].diff().fillna(0)
        
        # Extract time features
        df['Year'] = df[date_col].dt.year
        df['Month'] = df[date_col].dt.month
        df['Day'] = df[date_col].dt.day
        df['Day_of_week'] = df[date_col].dt.dayofweek
        df['Day_of_year'] = df[date_col].dt.dayofyear
    
    return df

start = 1
end = 3
date_col = 'ORDER_CREATION_DATE'
customer_col = 'unique_cust_id'
sales_col = 'amount_in_usd'

df = calculate_time_features(df, start, end, date_col, customer_col, sales_col)
df = df.sort_values(date_col)  # Sort DataFrame by 'ORDER_CREATION_DATE'
df = df.reset_index(drop=True)  # Reset index after sorting

df.head()
```

```python
def calculate_moving_averages(df, n):
    df[f'Last{N}-Day Sales Moving Average'] = df['Last_day_sales'].rolling(window=n).mean()
    
    df[f'Last{N}-Day Differences Moving Average'] = df['Last_day_diff'].rolling(window=n).mean()
    
    return df


N = 3  # Number of days for moving average calculation

df = calculate_moving_averages(df, N)
df.head()
```

```python
def calculate_last_month_sales(df, date_col, sales_col):
    # Convert the date column to datetime format
    df[date_col] = pd.to_datetime(df[date_col])
    
    # Calculate the start and end dates for the last month
    end_date = df[date_col].max()
    start_date = end_date - pd.DateOffset(months=1)
    
    # Filter the DataFrame for the last month sales
    last_month_sales = df[(df[date_col] >= start_date) & (df[date_col] <= end_date)]
    
    # Calculate the sum of sales for each customer
    last_month_sales = last_month_sales.groupby('CUSTOMER_NUMBER')[sales_col].sum().reset_index()
    
    # Merge the last month sales with the original DataFrame
    df = df.merge(last_month_sales, on='CUSTOMER_NUMBER', how='left')
    df.rename(columns={sales_col: 'last_month_sales'}, inplace=True)
    
    return df

date_col = 'ORDER_CREATION_DATE'
sales_col = 'amount_in_usd'

df = calculate_last_month_sales(df, date_col, sales_col)
df.head()
```
