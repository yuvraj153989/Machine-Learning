```python
def removehash(value):        #removing # from jersey number  
    value=value[1:]    #values will be fetched from 1st index..at 0th index there is # and after that there is number
    return int(value)
df['jersey']=df['jersey'].apply(removehash)
```

```python
df['salary']=df['salary'].apply(removehash)   #removal of $ symbol
```

```python
df['country'].value_counts()  #player from each country in dataset
```

```python
def remove_country_outlier(value):  #country=1 will be considered as others, for managing outliers
    if value not in ['USA','Canada','Australia']:
        return 'Others'
    else:
        return value
df['country']=df['country'].apply(remove_country_outlier)
```

```python
df['country'].value_counts()
```

```python
df['position'].value_counts()
```

```python
df['draft_round'].value_counts()
```

```python
def remove_undraft(value):    #removal of undrafted players
    if value=='Undrafted':
        return
    else:                    #undrafted ke jgh NULL aa jaayega using this fnctn
        return value
df['draft_round']=df['draft_round'].apply(remove_undraft)
df['draft_peak']=df['draft_peak'].apply(remove_undraft)
df=df.dropna()             #aaye hue NULL values ko remove krdo
```

```python
df.describe()
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le= LabelEncoder()
df['position']=le.fit_transform(df['position'])           #string labels to interger labels
df['country']=le.fit_transform(df['country'])
df['team']=le.fit_transform(df['team'])
df.head()
df=df.drop(['version'],axis=1)
```

```python
df.describe()
```

```python
df.shape
```

```python
x=df.drop(['salary'],axis=1)
y=df['salary']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
predictions=lr.predict(x_test)
```

```python
diff=y_test-predictions
```
